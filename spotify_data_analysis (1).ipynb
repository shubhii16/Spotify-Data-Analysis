# 🎧 Spotify Data Analysis

## ❓ Problem Statement

With the increasing popularity of music streaming platforms, it is important to understand what makes a song successful.

This project aims to answer:
- Which artists dominate the Spotify charts?
- What factors influence song popularity?
- Is there any relationship between song duration and popularity?
- How much explicit content is present in trending songs?
- How do music trends change over time?

The analysis helps in understanding patterns in global music consumption.

```python
## 📂 Data Source
spotify Dataset from kaggle
```

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

df = pd.read_csv('spotify-top-50-world.csv')

df.head()
```

```python
df.info()
```

```python
df['date'] = pd.to_datetime(df['date'])
df['release_date'] = pd.to_datetime(df['release_date'], errors='coerce')
```

```python
df.drop('album_cover_url', axis=1, inplace=True)
```

```python
df['duration_min'] = df['duration_ms'] / 60000
```

```python
df.head()
```

```python
top_artists = df['artist'].value_counts().head(10)
print(top_artists)
```

```python
top_artists.plot(kind='bar')
plt.title("Top 10 Artists")
plt.show()
```

```python
df.sort_values(by='popularity', ascending=False)[['song','artist','popularity']].head(10)
```

```python
df['is_explicit'].value_counts().plot(kind='pie', autopct='%1.1f%%')
plt.title("Explicit Content Distribution")
plt.show()
```

```python
sns.histplot(df['duration_min'], bins=20)
plt.title("Song Duration Distribution")
plt.show()
```

```python
df['album_type'].value_counts().plot(kind='bar')
plt.title("Album Type Distribution")
plt.show()
```

```python
sns.heatmap(df[['popularity','duration_ms','total_tracks']].corr(), annot=True)
plt.title("Correlation Matrix")
plt.show()
```

## 📊 Conclusion

After analyzing the Spotify dataset, the following insights were observed:

- A few artists dominate the global music charts.
- Most popular songs have a duration between 2 to 4 minutes.
- There is no strong relationship between song duration and popularity.
- Explicit songs make up a significant portion of the dataset.
- Music trends remain relatively stable with some fluctuations over time.

The analysis provides a clear understanding of modern music trends and listener behavior.
